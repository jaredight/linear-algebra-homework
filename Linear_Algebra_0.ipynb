```
def first_rpt(M):
  new_array=np.array(M)
  for i in range(len(M)):
    new_array[i]=M[0]
  return new_array
```

```
my_matrix=np.array([[1, 2, 3, 4],[-5, -6, -7, -8],[1, 5, 2, 3]])
print(first_rpt(my_matrix))
```

```
def matrix_sum(M):
  sum=0
  for i in range(len(M)):
    for j in range(len(M[0,:])):
      sum=sum+M[i,j]
  return sum
```

```
mat=np.array([[1,-1,2,-3,1,1],[-2,-2,0,1,1,-5],[1,1,1,1,-2,-1]])
print(matrix_sum(mat))
```
