**Problem 2**

```
def first_column_zeros(A):
  B=np.copy(A)
  for i in range(1,len(A)):
    B[i,]=B[i,:]-(B[i,0]/B[0,0])*B[0,:]
  return B

#A=np.array([[1,1,1],[1,4,2],[4,7,8],[34,5,-4]])
#print(first_column_zeros(A))
```

**Problem 3**

```
def row_echelon(A,b):  #row reduces the augmented matrix using only one type of row operation, adding a multiple of one row to another.
  M=augment(A,b)
  rows,cols = np.shape(M)
  for c in range(cols):
    for r in range(c+1,rows):
      M[r,:]=M[r,:]-(M[r,c]/M[c,c])*M[c,:]
  return M 

#A=np.array([[1,1,1,8],[1,4,2,5],[4,7,9,8],[2,3,4,4]])
#b=np.array([1,3,9,8])
#print(augment(A,b))
#print(row_echelon(A,b))
```

**Problem 4**

```
def LU_decomposition(A): #accepts a square matrix and factors as LU
  m,n=np.shape(A)
  U=np.copy(A)
  L=np.identity(m)
  #print(U)
  #print(L)
  for j in range(0,n):
    for i in range(j+1,m):
      L[i,j] = U[i,j]/U[j,j]
      U[i,:] = U[i,:] - L[i,j]*U[j,:]
  return L,U
#A=np.array([[3,1,-2],[1.5,2,-5],[2,-4,1]])
#L,U = LU_decomposition(A)
#print(L)
#print(U)
```

```
def back_substitution(U,b): #accepts an upper triangular square matrix U and a vector b, solves Ux=b for x.
  x=np.array([])
  x = np.linalg.inv(U).dot(b)
  return x

#U=np.array([[2,-3.1,1],[0,1,3],[0,0,4]])
#y=np.array([1,-2.1,3])
#print(back_substitution(U,y))
```

**Problem 6**

```
def forward_substitution(L,b): #accepts a lower triangular square matrix L and a vector b, solves Ly=b.
  y=np.array([])
  rows,cols=np.shape(L)
  for m in range(0,rows):
    rowsum=0
    for i in range(0,m):
      rowsum=rowsum+L[m,i]
    a=(b[m]-rowsum)/L[m,m]
    for j in range(0,rows):
      L[j,m]=L[j,m]*a
    y=np.append(y,a)
  return y

#L=np.array([[1,0,0],[3,1,0],[-1.1,2,1]])
#b=np.array([-2.1,1,-1])
#print(forward_substitution(L,b))
```
