**Problem 4**

```
from scipy import sparse
def sparse_laplace1D(n):
  diag1=np.ones(n-1)
  diag_4=-4*np.ones(n)
  sparse_matrix=sparse.diags([diag1,diag_4,diag1],[-1,0,1])
  return sparse_matrix 

#sparse_laplace1D(9).toarray()
```

**Problem 5**

```
def sparse_laplace(n):
  
  B = sparse_laplace1D(n)
  
  vec = np.ones(n)
  I = sparse.diags(vec)
  I.toarray()

  sparse_matrix = sparse.bmat([[B if i==j else I if abs(i-j)==1 else None 
                                for i in range(n)] for j in range(n)])
  return sparse_matrix 

#plt.spy(sparse_laplace(4))
```
