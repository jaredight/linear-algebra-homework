```
import numpy as np
from matplotlib import pyplot as plt
y = np.arange(-5,6)**2
print(y)
plt.plot(y) #Draw the line plot
plt.show() #Reveal the resulting plot
# Creates an array of 50 evenly spaced points between -5 and 5.
x = np.linspace(-5,5,50)
# Creates an array y which contains the squares of the values in x.
y=x**2
plt.plot(x,y)
plt.show()
```

```
from scipy import sparse
# Define the rows, columns, and values separately.
rows = np.array([0, 4, 0,0]) # the row index of the nonzero entries
cols = np.array([1, 2, 0,0]) # the column index of the nonzero entries
vals = np.array([1, 5, 2,9]) # the actual value at the nonzero entries
A = sparse.coo_matrix((vals, (rows,cols)), shape=(5,3))
print(A)
print(A.toarray())
plt.spy(A)
```

```
A = sparse.coo_matrix(np.ones([2,2]))
B = sparse.coo_matrix(np.full([2,2],2.))
# sparse.bmat() creates a block matrix. Use "None" for zero blocks.
C = sparse.bmat([[A,None,A],[None,B,None]])
print(C)
print(C.toarray())
D = sparse.block_diag((A,B,B))
print(D)
print(D.toarray())
```

```
# Construct and show a matrix with 50 2x3 diagonal blocks.
B = sparse.coo_matrix([[1,3,5],[7,9,11]])
A = sparse.block_diag([B]*50) #this makes 50 copies of the matrix B
plt.spy(A,markersize=3)
plt.show()
```

```
n = 6 #this could also be done for arbitrary n
vec =5* np.ones(n) #creates a vector of ones of size n, then multiplies by 5.
A = sparse.diags(vec)
A.toarray()
```

```
diag0 = 5*np.ones(7)
diag2 = -2*np.ones(7)
diag_1 = -1*np.ones(7)
diagonals = [diag0, diag2, diag_1] #the set of diagonal vectors
A = sparse.diags(diagonals,[0,2,-1])
A.toarray() #just to check, see if the matrix looks correct
```

```
diag0 = 5*np.ones(7)
diag2 = 2*np.ones(7)
diag_1 = 2*np.ones(7)
diagonals = [diag0, diag2, diag_1] #the set of diagonal vectors
A = sparse.diags(diagonals,[0,2,-1])

#Here we define B as the 7x7 identity matrix
n = 7 #this could also be done for arbitrary n
vec = np.ones(n) #creates a vector of ones of size n, then multiplies by 5.
B = sparse.diags(vec)
B.toarray()

D = sparse.bmat([[A if i==j else B if abs(i-j)==1 else None for i in range(2)]
                 for j in range(2)])
D.toarray()
#plt.spy(D)
```
