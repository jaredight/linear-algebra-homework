**Problem 1**

```
def adj_matrix(n,edge_list): 
  adj_m=np.zeros((n,n))
  for i in range(0,len(edge_list)):
    adj_m[edge_list[i,0],edge_list[i,1]]=adj_m[edge_list[i,0],edge_list[i,1]]+1
  return adj_m 

E1=np.array([[0,1],[1,4],[0,2],[0,4],[1,3],[2,0],[2,4],[3,4],[3,2],[3,2]])
adj_matrix(5,E1)
```

**Problem 2**

```
def degree_cent(n,edge_list):
  deg_array=np.zeros(n)
  adjmat=adj_matrix(n,edge_list)
  for i in range(len(adjmat)):
    for j in range(len(adjmat)):
      deg_array[i]=deg_array[i]+adjmat[j,i]
  return deg_array

E1=np.array([[0,1],[1,4],[0,2],[0,4],[1,3],[2,0],[2,4],[3,4],[3,2],[3,2]])
degree_cent(5,E1)
```
